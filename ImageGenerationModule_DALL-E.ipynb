```python
import openai
import os 
import requests 
from PIL import Image 
from io import BytesIO

openai.api_key = os.environ.get("OPENAI_API_KEY")
```

A Function for text-to-image generation using create endpoint of DALL-E API function takes in a string argument

text - describing the generated image

n - number of images to generate

size - size of each generated image

Return the URL of one image as we are generating only one image 

Note: The size of the generated images must be one of 256×256, 512×512, or 1024×1024.

```python
def generate(text):
  res = openai.Image.create(
    prompt=text,
    n=1, 
    size="256x256", 
  )
  return res["data"][0]["url"]
```
